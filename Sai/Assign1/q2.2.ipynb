```python
import numpy as np
from numpy import random
x=random.normal(size=1000,loc=50,scale=5)
print(x)
```

```python
print(np.mean(x))
```

```python
print(np.std(x))
```

```python
print(np.median(x))
```

```python
print(np.var(x))
```

```python
print(np.percentile(x,25))
```

```python
print(np.percentile(x,75))
```
