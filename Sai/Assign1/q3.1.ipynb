```python
import numpy as np
from numpy import random 
import matplotlib.pyplot as plt 

a=np.linspace(-10,10,num=100000)
```

```python
b=a*a
```

```python
plt.plot(a,b,color="r")
plt.show()
```

```python
u=random.uniform(size=100) #default low=0 and high=1
v=random.normal(size=100,loc=0.5,scale=0.1)
```

```python
plt.scatter(u,v)
plt.show()
```

```python
e=random.exponential(size=500) #default scale value=1.0
plt.hist(e)
plt.show()
```
