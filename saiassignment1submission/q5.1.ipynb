```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
h=np.arange(1,32)
```

```python
d=pd.Series(h)
```

```python
s=np.array([])
for x in h:
    s=np.append(s,15+(10*math.sin(2*math.pi*x/31)))   
```

```python
sin=pd.Series(s)
y=pd.date_range("01-01-26",periods=31)
df=pd.DataFrame({})
df["Dates"]=y
df["Temperatures"]=sin
print(df)
```

```python
m=np.array([])
i=1
while i<=22:
 m=np.append(m,sin[i:i+7].mean())
 i+=7
```

```python
plt.plot(d,sin)
plt.xlabel("Date in January")
plt.ylabel("Temperature")
plt.scatter([7,14,21,28],m,color="r") #rolling average values
plt.show()
```
