```python
#1.2
def isPrime(n):
    for i in range(2,int(n**(1/2))+1):
        if(n % i ==0):
            return False
    return True

n1 = int(input("Enter first number "))
n2 = int(input("Enter second number "))
for i in range(n1,n2+1):
    if(isPrime(i)):
        print(i)
```

```python
#4.1
import pandas as pd
dict1 = {"Name":['Alice', 'Bob', 'Charlie','Diana','Ethan'], "Age": [28,34,25,42,30], "Salary":[70000,80000,50000, 110000,75000]}
df1 = pd.DataFrame(dict1)
print(df1)
df1["Tax"] = df1["Salary"]*0.2
print(df1[df1["Age"]>=30])
sum_salary = 0
people = 0
for i in (df1[df1["Age"] < 30]["Salary"]):
    people = people + 1
    sum_salary += i
print("Mean salary is ",sum_salary/people)
df1.sort_values(by=["Salary"], ascending = False, inplace = True)
df1
```

```python
#5.1
df_temp = pd.date_range(start = "2025-01-01", periods = 31, freq= 'D')
print(df_temp)
x = np.linspace(0, 2*np.pi,31, endpoint=True)
temp = 15 + 10*np.sin(x)
plt.plot(df_temp, temp)
rolling_avg = 0
avg_temp = []
for i in range(7):
    rolling_avg+= temp[i]
rolling_avg /= 7
avg_temp.append(rolling_avg)
day = 7
while (day<31):
    rolling_avg = (rolling_avg * 7 - temp[day-7]+ temp[day])/7
    avg_temp.append(rolling_avg)
    day +=1 
plt.plot(df_temp[6:], avg_temp)
plt.show()
print(avg_temp)
print(df_temp[temp > 20])
```
